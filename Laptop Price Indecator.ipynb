```python
# Convert 'Memory' column to string and clean up the values
price_ind['Memory'] = price_ind['Memory'].astype(str).replace(r'\.0', '', regex=True)
price_ind['Memory'] = price_ind['Memory'].str.replace('GB', '', regex=False)
price_ind['Memory'] = price_ind['Memory'].str.replace('TB', '000', regex=False)

# Split the 'Memory' column into two parts based on the '+' symbol
new = price_ind['Memory'].str.split("+", n=1, expand=True)

# Create 'first' and 'second' columns from the split data
price_ind['first'] = new[0].str.strip()
price_ind['second'] = new[1].str.strip().fillna('0')  # Fill missing values in 'second' column with '0'

# Create indicator columns to identify memory types in 'first'
price_ind['Layer1HDD'] = price_ind['first'].apply(lambda x: 1 if 'HDD' in x else 0)
price_ind['Layer1SSD'] = price_ind['first'].apply(lambda x: 1 if 'SSD' in x else 0)
price_ind['Layer1Hybrid'] = price_ind['first'].apply(lambda x: 1 if 'Hybrid' in x else 0)
price_ind['Layer1Flash_Storage'] = price_ind['first'].apply(lambda x: 1 if 'Flash Storage' in x else 0)

# Remove non-numeric characters from 'first' and 'second' columns
price_ind['first'] = price_ind['first'].str.replace(r'\D', '', regex=True)
price_ind['second'] = price_ind['second'].str.replace(r'\D', '', regex=True)

# Convert 'first' and 'second' columns to integers, handling non-numeric values
price_ind['first'] = pd.to_numeric(price_ind['first'], errors='coerce').fillna(0).astype(int)
price_ind['second'] = pd.to_numeric(price_ind['second'], errors='coerce').fillna(0).astype(int)

# Create indicator columns to identify memory types in 'second'
price_ind['Layer2HDD'] = price_ind['second'].apply(lambda x: 1 if x > 0 else 0)
price_ind['Layer2SSD'] = price_ind['second'].apply(lambda x: 1 if x > 0 else 0)
price_ind['Layer2Hybrid'] = price_ind['second'].apply(lambda x: 1 if x > 0 else 0)
price_ind['Layer2Flash_Storage'] = price_ind['second'].apply(lambda x: 1 if x > 0 else 0)

# Calculate total values for each memory type
price_ind['HDD'] = (price_ind['first'] * price_ind['Layer1HDD']) + (price_ind['second'] * price_ind['Layer2HDD'])
price_ind['SSD'] = (price_ind['first'] * price_ind['Layer1SSD']) + (price_ind['second'] * price_ind['Layer2SSD'])
price_ind['Hybrid'] = (price_ind['first'] * price_ind['Layer1Hybrid']) + (price_ind['second'] * price_ind['Layer2Hybrid'])
price_ind['Flash_Storage'] = (price_ind['first'] * price_ind['Layer1Flash_Storage']) + (price_ind['second'] * price_ind['Layer2Flash_Storage'])

# Drop intermediate columns that are no longer needed
price_ind.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
                        'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
                        'Layer2Flash_Storage'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
```

```python
# Transforming the columns using OneHotEncoder for specified indices
column_transformer = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
    ],
    remainder='passthrough'
)

# Creating a linear regression model
linear_model = LinearRegression()

# Creating a pipeline with the column transformer and linear regression
model_pipeline = Pipeline(steps=[
    ('column_transform', column_transformer),
    ('linear_regression', linear_model)
])

# Fitting the pipeline to the training data
model_pipeline.fit(X_train, y_train)

# Making predictions on the test set
predictions = model_pipeline.predict(X_test)

# Displaying the R2 score and Mean Absolute Error
print('R-squared:', r2_score(y_test, predictions))
print('Mean Absolute Error:', mean_absolute_error(y_test, predictions))
```

```python
from sklearn.ensemble import VotingRegressor,StackingRegressor

# Step 1: Preprocessing the data with OneHotEncoder
step1 = ColumnTransformer(
    transformers=[
        ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
    ],
    remainder='passthrough'
)

# Step 2: Initialize the individual regressors with bootstrap set to True
rf = RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, 
                           max_features=0.75, max_depth=15, bootstrap=True)
gbdt = GradientBoostingRegressor(n_estimators=100, max_features=0.5)
xgb = XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5)
et = ExtraTreesRegressor(n_estimators=100, random_state=3, max_samples=0.5, 
                         max_features=0.75, max_depth=10, bootstrap=True)

# Step 3: Create the Voting Regressor
step2 = VotingRegressor(estimators=[('rf', rf), ('gbdt', gbdt), ('xgb', xgb), ('et', et)],
                        weights=[5, 1, 1, 1])

# Step 4: Build the pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Step 5: Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Step 6: Predict the target for the test set
y_pred = pipe.predict(X_test)

# Step 7: Evaluate the model
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```

```python
from sklearn.ensemble import VotingRegressor,StackingRegressor

step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')


estimators = [
    ('rf', RandomForestRegressor(n_estimators=350,random_state=3,max_samples=0.5,max_features=0.75,max_depth=15)),
    ('gbdt',GradientBoostingRegressor(n_estimators=100,max_features=0.5)),
    ('xgb', XGBRegressor(n_estimators=25,learning_rate=0.3,max_depth=5))
]

step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
